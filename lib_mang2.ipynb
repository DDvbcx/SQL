#### 练习11

```python
query_fifty = '''SELECT b.BookCategory, COUNT(*) AS totalcount, b.BookID
                 FROM ReaderInfo AS r
                 LEFT OUTER JOIN BookInfo AS b
                 ON r.BookID = b.BookID
                 WHERE r.BorrowCount >=2
                 GROUP BY b.BookCategory, b.BookID; '''
cursor.execute(query_fifty)
res = cursor.fetchall()
df_fifty11 = pd.DataFrame(res, columns=cursor.column_names)
df_fifty11
```
