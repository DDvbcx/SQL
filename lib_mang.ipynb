#### 练习7

```python
### 连接查询
query_bt = '''SELECT 
              r.Name,
              r.Phone
              FROM ReaderInfo AS r
              LEFT OUTER JOIN BookInfo AS b
              ON r.BookID = b.BookID
              WHERE b.Book_Name='Pride and Prejudice'; '''
cursor.execute(query_bt)

res = cursor.fetchall()
df_query7 = pd.DataFrame(res, columns=cursor.column_names)
df_query7
```

```python
### 嵌套查询
query_bt2 = '''SELECT Name, Phone
               FROM ReaderInfo 
               WHERE BookID in (SELECT BookID
                                FROM BookInfo
                                WHERE Book_Name='Pride and Prejudice'); '''
cursor.execute(query_bt2)

res = cursor.fetchall()
df_query7_ = pd.DataFrame(res, columns=cursor.column_names)
df_query7_
```
